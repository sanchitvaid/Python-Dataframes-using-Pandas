## MISSING VALUE TREATMENT

```python
df1  = pd.read_csv('class-grades.csv')
df1
```

```python
#FINDING COLUMNS WITH NULL VALUES

df1.isna().any()
```

```python
#IMPORTING LIBRARIES TO IMPUTE MEAN VALUE

from sklearn.impute import SimpleImputer
```

```python
mean_imputer = SimpleImputer(missing_values = np.nan , strategy = 'mean')
mean_imputer = mean_imputer.fit(df1)
imputed_df1 = mean_imputer.transform(df1.values)


#CONVERTING THE IMPUTED DF1 INTO TO A DF AND ASSIGNING THE ORIGINAL DATA,COLUMN VALUES TO IT

new_df1 = pd.DataFrame( data = imputed_df1 , columns = ['Prefix','Assignment','Tutorial','Midterm','TakeHome','Final'])
new_df1
```

```python
new_df1.isna().any()
```

```python
#IMPUTING NULL VALUES WITH MEDIAN VALUES

median_imputer = SimpleImputer(missing_values = np.nan , strategy = 'median')
median_imputer = median_imputer.fit(df1)

imputed_df1 = median_imputer.transform(df1.values)

new_df2 = pd.DataFrame(imputed_df1 , columns = ['Prefix','Assignment','Tutorial','Midterm','TakeHome','Final'])
new_df2
```
